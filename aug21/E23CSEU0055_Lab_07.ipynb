```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
```

1.	Data Loading and Preprocessing:
•	Load the dataset.

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv")
print(df)
```

•	Identify features (independent variables) and target (dependent variable).

```python
x = df[["acceleration"]]
y = df[["mpg"]].values
```

3.	Training and Prediction:
<br>
•	Split the dataset into training and testing sets.
<br>
•	Train both models on the training data.
<br>
•	Predict outputs on the test set.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
```

2.	Model Implementation:
<br>
•	Implement Linear Regression from scratch or using libraries.
<br>
•	Implement Polynomial Regression (choose polynomial degree 2 or 3 for comparison).

```python
l_red = LinearRegression()
l_red.fit(x_train, y_train)
yl_pred = l_red.predict(x_test)
```

```python
poly2 = PolynomialFeatures(degree=2)
x_poly2 = poly2.fit_transform(x_train)
xt_poly2 = poly2.fit_transform(x_test)
pr2 = LinearRegression()
pr2.fit(x_poly2, y_train)
yp2_pred = pr2.predict(xt_poly2)
```

```python
poly3 = PolynomialFeatures(degree=3)
x_poly3 = poly3.fit_transform(x_train)
xt_poly3 = poly3.fit_transform(x_test)
pr3 = LinearRegression()
pr3.fit(x_poly3, y_train)
yp3_pred = pr3.predict(xt_poly3)
```

4.	Performance Metrics Calculation:
<br>
Calculate and compare the following metrics for both models:
<br>
•	R² (Coefficient of Determination): Measures how well the regression predictions approximate the real data points. <br>
•	MAE (Mean Absolute Error): Measures the average magnitude of errors in predictions, without considering their direction.


```python
print("R-square Score")
print(f"R2 - Linear --> {r2_score(y_test, yl_pred)}")
print(f"R2 - Poly - Deg 2 --> {r2_score(y_test, yp2_pred)}")
print(f"R2 - Poly - Deg 3 --> {r2_score(y_test, yp3_pred)}")
```

```python
print("Mean Absolute Error")
print(f"MAE - Linear --> {mean_absolute_error(y_test, yl_pred)}")
print(f"MAE - Poly - Deg 2 --> {mean_absolute_error(y_test, yp2_pred)}")
print(f"MAE - Poly - Deg 3 --> {mean_absolute_error(y_test, yp3_pred)}")
```

5.	Plotting:
<br>
•	Plot the data points along with both regression lines/curves. <br>
•	Visualize and interpret the fit of each model.


```python
x_range = np.linspace(x.min(), x.max(), 200).reshape(-1, 1)
y_lin_line = l_red.predict(x_range)
plt.scatter(x, y, color="gray", alpha=0.5, label="Data Points")
plt.plot(x_range, y_lin_line, color="blue", label="Linear Regression")

y_poly2_line = pr2.predict(poly2.transform(x_range))
plt.plot(x_range, y_poly2_line, color="red", label="Polynomial (Degree 2)")

y_poly3_line = pr3.predict(poly3.transform(x_range))
plt.plot(x_range, y_poly3_line, color="green", label="Polynomial (Degree 3)")

plt.xlabel("Acceleration")
plt.ylabel("MPG")
plt.title("Linear vs Polynomial Regression")
plt.legend()
plt.show()
```

# Discussion

1. Linear Regression
Assumes a straight-line relationship between acceleration and mpg.<br>
Simpler, faster, but may underfit if the true relationship is curved.
<hr>

2. Polynomial Regression <br>
Degree 2/3 fits curves, capturing non-linear relationships better. <br>
Generally gives higher R² and lower MAE than linear regression.
<hr>

3. When to prefer Polynomial Regression <br>
When the data shows curved trends instead of a straight-line relationship. <br>
Useful when linear regression underfits.
<hr>

4. Overfitting vs Underfitting <br>
Underfitting: Linear regression oversimplifies the relationship, failing to capture the actual curvature between horsepower and acceleration. <br>
Better fit: Polynomial regression (degree 2) adapts to the curve, improving predictive power. <br>
Overfitting risk: Higher-degree polynomial (3+) might fit noise rather than the true relationship, especially since acceleration depends on multiple factors not modeled here.
